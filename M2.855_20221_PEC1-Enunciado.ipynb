Usando el conjunto de datos original:
- Definid un modelo Random Forest (fijando `n_estimators=10` para mantener el modelo simple y `random_state=seed`).
- Aplicad validación cruzada con el modelo definido y el dataset de train (con `cv=5` ya es suficiente).
- Como métrica para cuantificar el error del modelo, utilizad tanto MAE (la media del error absoluto), MSE (error cuadrático medio) y MAPE (porcentaje de error absoluto medio), separadamente.
- Calculad la media y la desviación estándard del error obtenido en cada caso.

**Notas**:
- Para entrenar un modelo random forest podéis usar `RandomForestRegressor` de `sklearn`.

```python
# Se define el dataset de variables y clase con los datos originales
X = pd.DataFrame(data['data'], columns = data['feature_names'])
y = data['target']
print(f"Dimensiones de X (vars):{X.shape}\nDimensiones de y (target): {y.shape}")

# Definición del modelo
model = RandomForestRegressor(n_estimators=10, random_state=seed)
```

Cuantificación del error utilizando MAE (_Mean Absolute Error_)

```python
mae_scores = cross_val_score(model, X, y, cv=5, scoring="neg_mean_absolute_error")
print(f"MAE Score: {np.abs(np.mean(mae_scores)):.4f} y la desviación estándar {np.std(mae_scores):.4f}")
```

Cuantificación del error utilizando MSE (_Mean Squared Error_)

```python
mse_scores = cross_val_score(model, X, y, cv=5, scoring="neg_mean_squared_error")
print(f"MSE Score: {np.abs(np.mean(mse_scores)):.4f} y la desviación estándar {np.std(mse_scores):.4f}")
```

Cuantificación del error utilizando MAPE (Mean Absolute Percentage Error)

```python
mape_scores = cross_val_score(model, X, y, cv=5, scoring="neg_mean_absolute_percentage_error")
print(f"MAPE Score: {round(np.abs(np.mean(mape_scores)),4)} y la desviación estándar {np.std(mape_scores):.4f}")
```

<div style="background-color: #EDF7FF; border-color: #7C9DBF; border-left: 5px solid #7C9DBF; padding: 0.5em;">
<strong>Análisis</strong>
</div>

¿A qué se deben las diferencias numéricas entre las distintas métricas?

¿Qué muestras y errores crees que pueden influir más o menos en el incremento/decremento de las distintas métricas?

Las diferencias son distintas porque las fórmulas de calcularlos son distintas y cada una de ellas toma en consideración aspectos diferentes.

- MAE mide el error de las diferencias absolutas entre los valores objetivos y las predicciones obtenidas. Todas las diferencias individuales tienen el mismo peso en la media pueto que es una puntuación lineal.

- MSE mide el error cuadrático medio de los resultados y luego para cada punto calcula el cuadrado de la diferencia entre las predicciones y el objetivo, promediándolos todos, potenciando al cuadrado el resultado de cada error.

- MAPE es la diferencia de error relativo, dividiendo cada error absoluto entre el valor objetivo, pudiendo considerarse una versión ponderada de MAE.

Fuente: https://sitiobigdata.com/2018/08/27/machine-learning-metricas-regresion-mse/

Como MSE aumenta la diferencia elevando al cuadrado los errores, es coherente con el resultado obtenido puesto que el dataset presenta una alta varianza. Sin embargo aplicando MAE los valores ya son más razonables puesto que es una variación lineal, como se especifica en la definición, aunque en MAPE, ponderando los valores, preenta los resultados más prometedores.

Para este último caso se ha dividido el dataset en train y test, aplicándose luego Cross Validation sobre el dataset de Train.

```python
# Proporción de separación de datasets de train y test
train_perc = 0.7

# Separación de dataset de train y test, en variables de decisión y de clase
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=(1-train_perc))
print(f"Dimensiones de Train\n--------------------\n\tX (vars):{X_train.shape}\n\ty (target): {y_train.shape}\n")
print(f"Dimensiones de Test\n--------------------\n\tX (vars):{X_test.shape}\n\ty (target): {y_test.shape}")
```

</div>

- Definid un modelo Random Forest (fijando `n_estimators=10` para mantener el modelo simple y `random_state=seed`).

- Como métrica para cuantificar el error del modelo, utilizad tanto accuracy como roc_auc.

```python
# Se define el modelo y se entrena con los datos de train
model = RandomForestClassifier(n_estimators=10, random_state=seed)
```
